# PrintShapes — Рисование фигур на C#
Каждая функция `GetX(...)` возвращает массив строк, `PrintX(...)` выводит фигуру в консоль.

```
// ============================================================
// ПРЯМОУГОЛЬНИК
// ============================================================

string[] GetRectangle(int width, int height, char symbol = '*', bool filled = false, int borderThickness = 1)
{
    var lines = new string[height];
    for (int row = 0; row < height; row++)
    {
        var line = new System.Text.StringBuilder();
        for (int col = 0; col < width; col++)
        {
            bool onBorder = row < borderThickness || row >= height - borderThickness
                         || col < borderThickness || col >= width - borderThickness;
            line.Append(filled || onBorder ? symbol : ' ');
        }
        lines[row] = line.ToString();
    }
    return lines;
}

void PrintRectangle(int width, int height, char symbol = '*', bool filled = false, int borderThickness = 1)
{
    foreach (var line in GetRectangle(width, height, symbol, filled, borderThickness))
        Console.WriteLine(line);
}

// Тесты
Console.WriteLine("=== Прямоугольник (контур) ===");
PrintRectangle(10, 5);

Console.WriteLine("\n=== Прямоугольник (залитый) ===");
PrintRectangle(10, 5, '#', filled: true);

Console.WriteLine("\n=== Прямоугольник (рамка толщиной 2) ===");
PrintRectangle(12, 6, '+', filled: false, borderThickness: 2);
```

```
// ============================================================
// РАВНОБЕДРЕННЫЙ ПРЯМОУГОЛЬНЫЙ ТРЕУГОЛЬНИК
// ============================================================

// direction: "right" — прямой угол снизу-слева
//            "left"  — прямой угол снизу-справа
string[] GetTriangle(int size, char symbol = '*', bool filled = false, string direction = "right")
{
    var lines = new string[size];
    for (int row = 0; row < size; row++)
    {
        int count = row + 1;
        var line = new System.Text.StringBuilder();
        if (direction == "left")
        {
            line.Append(new string(' ', size - count));
            for (int col = 0; col < count; col++)
            {
                bool onBorder = row == size - 1 || col == count - 1 || col == 0;
                line.Append(filled || onBorder ? symbol : ' ');
            }
        }
        else
        {
            for (int col = 0; col < count; col++)
            {
                bool onBorder = row == size - 1 || col == count - 1 || col == 0;
                line.Append(filled || onBorder ? symbol : ' ');
            }
        }
        lines[row] = line.ToString();
    }
    return lines;
}

void PrintTriangle(int size, char symbol = '*', bool filled = false, string direction = "right")
{
    foreach (var line in GetTriangle(size, symbol, filled, direction))
        Console.WriteLine(line);
}

Console.WriteLine("=== Треугольник (контур, вправо) ===");
PrintTriangle(6);

Console.WriteLine("\n=== Треугольник (залитый, влево) ===");
PrintTriangle(6, '#', filled: true, direction: "left");
```

```
// ============================================================
// ПЕСОЧНЫЕ ЧАСЫ
// ============================================================

string[] GetHourglass(int size, char symbol = '*', bool filled = false)
{
    int rows = size * 2 - 1;
    var lines = new string[rows];
    for (int row = 0; row < rows; row++)
    {
        int half = row < size ? row : rows - 1 - row;
        int spaces = half;
        int count  = size * 2 - 1 - half * 2;
        var line   = new System.Text.StringBuilder();
        line.Append(new string(' ', spaces));
        for (int col = 0; col < count; col++)
        {
            bool onBorder = col == 0 || col == count - 1
                         || row == 0 || row == rows - 1;
            line.Append(filled || onBorder ? symbol : ' ');
        }
        lines[row] = line.ToString();
    }
    return lines;
}

void PrintHourglass(int size, char symbol = '*', bool filled = false)
{
    foreach (var line in GetHourglass(size, symbol, filled))
        Console.WriteLine(line);
}

Console.WriteLine("=== Песочные часы (контур) ===");
PrintHourglass(6);

Console.WriteLine("\n=== Песочные часы (залитые) ===");
PrintHourglass(5, '@', filled: true);
```

```
// ============================================================
// РОМБ
// ============================================================

string[] GetDiamond(int size, char symbol = '*', bool filled = false)
{
    int rows = size * 2 - 1;
    var lines = new string[rows];
    for (int row = 0; row < rows; row++)
    {
        int half   = row < size ? row : rows - 1 - row;
        int spaces = size - 1 - half;
        int count  = half * 2 + 1;
        var line   = new System.Text.StringBuilder();
        line.Append(new string(' ', spaces));
        for (int col = 0; col < count; col++)
        {
            bool onBorder = col == 0 || col == count - 1
                         || row == 0 || row == rows - 1;
            line.Append(filled || onBorder ? symbol : ' ');
        }
        lines[row] = line.ToString();
    }
    return lines;
}

void PrintDiamond(int size, char symbol = '*', bool filled = false)
{
    foreach (var line in GetDiamond(size, symbol, filled))
        Console.WriteLine(line);
}

Console.WriteLine("=== Ромб (контур) ===");
PrintDiamond(5);

Console.WriteLine("\n=== Ромб (залитый) ===");
PrintDiamond(5, 'O', filled: true);
```

```
// ============================================================
// ЗЕБРА (горизонтальные полосы с заданной толщиной)
// ============================================================

string[] GetZebra(int width, int height, char symbol = '*', char emptySymbol = ' ', int lineThickness = 1)
{
    var lines = new string[height];
    for (int row = 0; row < height; row++)
    {
        bool isStripe = (row / lineThickness) % 2 == 0;
        lines[row] = new string(isStripe ? symbol : emptySymbol, width);
    }
    return lines;
}

void PrintZebra(int width, int height, char symbol = '*', char emptySymbol = ' ', int lineThickness = 1)
{
    foreach (var line in GetZebra(width, height, symbol, emptySymbol, lineThickness))
        Console.WriteLine(line);
}

Console.WriteLine("=== Зебра (толщина 1) ===");
PrintZebra(12, 8);

Console.WriteLine("\n=== Зебра (толщина 2) ===");
PrintZebra(12, 10, '#', '-', lineThickness: 2);
```

```
// ============================================================
// ЗМЕЙКА
// ============================================================
// direction: "horizontal-cw"  — горизонтальная по часовой
//            "horizontal-ccw" — горизонтальная против часовой
//            "vertical-cw"    — вертикальная по часовой
//            "vertical-ccw"   — вертикальная против часовой

string[] GetSnake(int width, int height, int thickness = 2, char symbol = '*', string direction = "horizontal-cw")
{
    char[,] grid = new char[height, width];
    for (int r = 0; r < height; r++)
        for (int c = 0; c < width; c++)
            grid[r, c] = ' ';

    if (direction.StartsWith("horizontal"))
    {
        bool clockwise = direction == "horizontal-cw";
        int step = thickness;
        int row = clockwise ? 0 : height - step;
        bool goRight = true;

        while (row >= 0 && row + step <= height)
        {
            for (int r = row; r < row + step; r++)
                for (int c = 0; c < width; c++)
                    grid[r, c] = symbol;

            // соединитель
            int nextRow = clockwise ? row + step * 2 - step : row - step;
            if ((clockwise && nextRow + step <= height) || (!clockwise && nextRow >= 0))
            {
                int connCol = goRight ? width - step : 0;
                for (int r = row + (clockwise ? step : -step); clockwise ? r < row + step * 2 : r > row - step; r += clockwise ? 1 : -1)
                    for (int cc = connCol; cc < connCol + step && cc < width; cc++)
                        if (r >= 0 && r < height) grid[r, cc] = symbol;
            }

            row += clockwise ? step * 2 : -step * 2;
            goRight = !goRight;
        }
    }
    else // vertical
    {
        bool clockwise = direction == "vertical-cw";
        int step = thickness;
        int col = clockwise ? 0 : width - step;
        bool goDown = true;

        while (col >= 0 && col + step <= width)
        {
            for (int c = col; c < col + step; c++)
                for (int r = 0; r < height; r++)
                    grid[r, c] = symbol;

            int nextCol = clockwise ? col + step * 2 : col - step;
            if ((clockwise && nextCol + step <= width) || (!clockwise && nextCol >= 0))
            {
                int connRow = goDown ? height - step : 0;
                for (int r2 = connRow; r2 < connRow + step; r2++)
                    for (int c2 = col; c2 < col + step * 2 && c2 < width; c2++)
                        if (r2 >= 0 && r2 < height) grid[r2, c2] = symbol;
            }

            col += clockwise ? step * 2 : -step * 2;
            goDown = !goDown;
        }
    }

    var lines = new string[height];
    for (int r = 0; r < height; r++)
    {
        var sb = new System.Text.StringBuilder();
        for (int c = 0; c < width; c++) sb.Append(grid[r, c]);
        lines[r] = sb.ToString();
    }
    return lines;
}

void PrintSnake(int width, int height, int thickness = 2, char symbol = '*', string direction = "horizontal-cw")
{
    foreach (var line in GetSnake(width, height, thickness, symbol, direction))
        Console.WriteLine(line);
}

Console.WriteLine("=== Змейка горизонтальная ===");
PrintSnake(20, 12, thickness: 2);

Console.WriteLine("\n=== Змейка вертикальная ===");
PrintSnake(20, 12, thickness: 2, direction: "vertical-cw");
```

```
// ============================================================
// ВЛОЖЕННЫЕ ПРЯМОУГОЛЬНИКИ
// ============================================================

string[] GetNestedRectangles(int width, int height, char symbol = '*', int gap = 1)
{
    char[,] grid = new char[height, width];
    for (int r = 0; r < height; r++)
        for (int c = 0; c < width; c++)
            grid[r, c] = ' ';

    int level = 0;
    int top = 0, bottom = height - 1, left = 0, right = width - 1;

    while (top <= bottom && left <= right)
    {
        for (int c = left; c <= right; c++) { grid[top, c] = symbol; grid[bottom, c] = symbol; }
        for (int r = top; r <= bottom; r++) { grid[r, left] = symbol; grid[r, right] = symbol; }

        top    += gap + 1;
        bottom -= gap + 1;
        left   += gap + 1;
        right  -= gap + 1;
        level++;
    }

    var lines = new string[height];
    for (int r = 0; r < height; r++)
    {
        var sb = new System.Text.StringBuilder();
        for (int c = 0; c < width; c++) sb.Append(grid[r, c]);
        lines[r] = sb.ToString();
    }
    return lines;
}

void PrintNestedRectangles(int width, int height, char symbol = '*', int gap = 1)
{
    foreach (var line in GetNestedRectangles(width, height, symbol, gap))
        Console.WriteLine(line);
}

Console.WriteLine("=== Вложенные прямоугольники (gap=1) ===");
PrintNestedRectangles(21, 11);

Console.WriteLine("\n=== Вложенные прямоугольники (gap=0) ===");
PrintNestedRectangles(21, 11, '#', gap: 0);
```

```
// ============================================================
// УЛИТКА
// ============================================================
// direction: "cw"  — по часовой стрелке (снаружи внутрь)
//            "ccw" — против часовой

string[] GetSnail(int width, int height, char symbol = '*', string direction = "cw")
{
    char[,] grid = new char[height, width];
    for (int r = 0; r < height; r++)
        for (int c = 0; c < width; c++)
            grid[r, c] = ' ';

    int top = 0, bottom = height - 1, left = 0, right = width - 1;

    if (direction == "cw")
    {
        while (top <= bottom && left <= right)
        {
            for (int c = left; c <= right; c++)  grid[top, c]    = symbol; top++;
            for (int r = top; r <= bottom; r++)   grid[r, right]  = symbol; right--;
            if (top <= bottom) { for (int c = right; c >= left; c--) grid[bottom, c] = symbol; bottom--; }
            if (left <= right) { for (int r = bottom; r >= top; r--) grid[r, left]   = symbol; left++; }
        }
    }
    else // ccw
    {
        while (top <= bottom && left <= right)
        {
            for (int c = right; c >= left; c--)  grid[top, c]    = symbol; top++;
            for (int r = top; r <= bottom; r++)   grid[r, left]   = symbol; left++;
            if (top <= bottom) { for (int c = left; c <= right; c++) grid[bottom, c] = symbol; bottom--; }
            if (left <= right) { for (int r = bottom; r >= top; r--) grid[r, right]  = symbol; right--; }
        }
    }

    var lines = new string[height];
    for (int r = 0; r < height; r++)
    {
        var sb = new System.Text.StringBuilder();
        for (int c = 0; c < width; c++) sb.Append(grid[r, c]);
        lines[r] = sb.ToString();
    }
    return lines;
}

void PrintSnail(int width, int height, char symbol = '*', string direction = "cw")
{
    foreach (var line in GetSnail(width, height, symbol, direction))
        Console.WriteLine(line);
}

Console.WriteLine("=== Улитка по часовой ===");
PrintSnail(20, 10);

Console.WriteLine("\n=== Улитка против часовой ===");
PrintSnail(20, 10, '#', direction: "ccw");
```
